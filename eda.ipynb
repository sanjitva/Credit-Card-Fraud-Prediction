## Import Packages

```python
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
fraud_df = df[df['Class']== 1]
normal_df = df[df['Class']== 0]
```

## Fraud

```python
fraud_count = len(fraud_df)
normal_count = len(normal_df)
```

```python
# Identifying Fraud

def column_hist(column):

    fig, ax = plt.subplots(1,2,figsize=(48,20))

    ax[0].hist(normal_df[column], bins = 100, edgecolor='black')

    ax[1].hist(fraud_df[column], bins = 100, edgecolor='black')

    ax[0].set_title(f'{column} Normal',fontsize= 20)
    ax[1].set_title(f'{column} Fraud',fontsize= 20);
```

```python
warnings.filterwarnings("ignore")

columns = list(df.drop('Class',axis=1).columns)

for a in columns:
    column_hist(a)
```
